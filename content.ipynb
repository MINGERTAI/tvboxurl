```python
import requests
import json

def is_response_json(response):
    """
    判断给定的HTTP响应内容是否为有效的JSON格式。

    参数:
        response (requests.Response): HTTP响应对象

    返回:
        bool: 如果响应内容是JSON格式，返回True；否则返回False
    """
    try:
        json.loads(response.text)
        return True
    except json.JSONDecodeError:
        return False

# 示例请求
url = "http://lige.unaux.com/?url=http://rihou.vip:66/"
response = requests.get(url)

if is_response_json(response):
    print("网页返回的数据是JSON格式")
    print(response.text)
else:
    print("网页返回的数据不是JSON格式")
```
